```python
# Beispielcode für das Testen einer Request (ging ja nicht immer bei allen Seiten)

import requests # Anfrage an die Webseite

def get_response(url):
    timeout_seconds = 5  # Timeout auf 5 Sekunden setzen

    try:
        response = requests.get(url, timeout=timeout_seconds)

        # Status Code check
        if response.status_code == 200:
            # Überprüfen, ob Inhalt vorhanden ist
            if response.content:
                print("Inhalt erfolgreich geladen.")
                print(response.text[:500])  # Optional: Erste 500 Zeichen des Inhalts ausgeben
                return True
            else:
                print("Kein Inhalt vorhanden.")
        else:
            print(f"Fehler: Status Code ist {response.status_code}")

    except requests.Timeout:
        print(f"Timeout nach {timeout_seconds} Sekunden.")
    except requests.ConnectionError:
        print("Verbindungsfehler. Keine Verbindung zum Server.")
    except requests.RequestException as e:
        # Fängt alle anderen Request-Fehler ab
        print(f"Ein Fehler ist aufgetreten: {e}")

positiv_resp = []

for liste in liste_arbeitstellen:
    resp = get_response(liste)
    if resp == True:  # 200 steht für "OK" in HTTP
        positiv_resp.append(liste)
    else: 
        print(liste, 'negative')
        
print(positiv_resp)
```

```python
# Beispielcode um das zu umgehen

# Zuerst kopiert man via Inspector den ganzen code raus:
with open('htmltesttext.txt', 'r') as file:
    html2 = file.read()  # Den gesamten Inhalt der Datei in eine Variable speichern
    
from bs4 import BeautifulSoup
import json

def extract_all_job_details(html: str) -> str:
    # Parse the HTML
    soup = BeautifulSoup(html, 'html.parser')

    # Find all relevant job cards
    job_cards = soup.find_all('li', class_='css-5lfssm')

    if not job_cards:
        return json.dumps({"error": "No job cards found"}, indent=4, ensure_ascii=False)

    all_job_details = []

    for job_card in job_cards:
        try:
            # Extract job details
            job_title = job_card.find('span', {'id': lambda x: x and x.startswith('jobTitle')}).get_text(strip=True)
            company_name = job_card.find('span', {'data-testid': 'company-name'}).get_text(strip=True)
            location = job_card.find('div', {'data-testid': 'text-location'}).get_text(strip=True)
            rating = job_card.find('span', {'data-testid': 'holistic-rating'}).get_text(strip=True)
            job_type = job_card.find('div', {'data-testid': 'attribute_snippet_testid'}).get_text(strip=True)
            posted_date = job_card.find('span', {'data-testid': 'myJobsStateDate'}).get_text(strip=True)
            #.find('div').find('p')

            # Extract job description points
            job_description_items = job_card.find_all('li')
            job_description = [item.get_text(strip=True) for item in job_description_items]

            # Structure the extracted data
            job_details = {
                "job_title": job_title,
                "company_name": company_name,
                "location": location,
                "rating": rating,
                "job_type": job_type,
                "posted_date": posted_date,
                "job_description": job_description
            }

            all_job_details.append(job_details)

        except AttributeError:
            # Handle cases where a job card is incomplete
            all_job_details.append({"error": "Failed to parse some job details for this card"})

    # Convert all job details to JSON
    return json.dumps(all_job_details, indent=4, ensure_ascii=False)

# Beispielaufruf der Funktion
result = extract_all_job_details(html)
print(result)
```

```python
import sqlite3
from bs4 import BeautifulSoup
import json

# Funktion zur Extraktion und Speicherung der Jobdetails in eine SQLite-Datenbank
def extract_and_store_job_details_in_db(html: str, db_path: str):
    # Parse the HTML
    soup = BeautifulSoup(html, 'html.parser')

    # Find all relevant job cards
    job_cards = soup.find_all('li', class_='css-5lfssm')

    if not job_cards:
        print("No job cards found")
        return

    # Verbindung zur SQLite-Datenbank herstellen (wird erstellt, falls sie nicht existiert)
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()

    # Tabelle erstellen, falls sie noch nicht existiert
    cursor.execute('''CREATE TABLE IF NOT EXISTS jobs (
                        id INTEGER PRIMARY KEY AUTOINCREMENT,
                        job_title TEXT,
                        company_name TEXT,
                        location TEXT,
                        rating TEXT,
                        job_type TEXT,
                        posted_date TEXT,
                        job_description TEXT
                    )''')

    for job_card in job_cards:
        try:
            # Extract job details
            job_title = job_card.find('span', {'id': lambda x: x and x.startswith('jobTitle')}).get_text(strip=True)
            company_name = job_card.find('span', {'data-testid': 'company-name'}).get_text(strip=True)
            location = job_card.find('div', {'data-testid': 'text-location'}).get_text(strip=True)
            rating = job_card.find('span', {'data-testid': 'holistic-rating'}).get_text(strip=True)
            job_type = job_card.find('div', {'data-testid': 'attribute_snippet_testid'}).get_text(strip=True)
            posted_date = job_card.find('span', {'data-testid': 'myJobsStateDate'}).get_text(strip=True)

            # Extract job description points
            job_description_items = job_card.find_all('li')
            job_description = "\n".join([item.get_text(strip=True) for item in job_description_items])

            # Jobdetails in die Datenbank einfügen
            cursor.execute('''INSERT INTO jobs (job_title, company_name, location, rating, job_type, posted_date, job_description)
                              VALUES (?, ?, ?, ?, ?, ?, ?)''', 
                           (job_title, company_name, location, rating, job_type, posted_date, job_description))
        
        except AttributeError:
            print("Failed to parse job card")

    # Änderungen speichern und Verbindung schließen
    conn.commit()
    conn.close()

# Beispielaufruf der Funktion
db_path = 'jobs.db'  # Pfad zur SQLite-Datenbank
extract_and_store_job_details_in_db(html, db_path)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

def scatterplot(x_array, y_array, title = 'Scatterplot', x_title = 'x-achse', y_title = 'y-achse'):
    x = np.array(x_array)
    y = np.array(y_array)
    # Calculate variances
    variance_x = np.var(x)
    variance_y = np.var(y)

    # Print variance
    print("Variance of job counts:", variance_x)
    print("Variance of population:", variance_y)

    # Scatter plot
    plt.scatter(x, y, color='blue', label='Data Points')

    # Calculate the regression line
    slope, intercept, r_value, p_value, std_err = linregress(x, y)
    regression_line = slope * x + intercept

    # Plot the regression line
    plt.plot(x, regression_line, color='red', label=f'Regression Line (R² = {r_value**2:.2f})')

    # Labels and title
    plt.xlabel(x_title)
    plt.ylabel(y_title)
    plt.title(title)
    plt.legend()
    plt.show()
```

```python
# levenshtein distanz
# import editdistance

# def levenshtein_distance(word1, word2):
#     return editdistance.eval(word1, word2)

# word1 = "example"
# word2 = "samples"
# print(levenshtein_distance(word1, word2))  

import Levenshtein

# Sample keyword library of common job titles or keywords
job_keywords = [
    "Softwareentwickler", "Projektmanager", "Informatiker", "Mathematiker", 
    "Data Scientist", "Analyst", "Ingenieur", "Berater", "Entwickler", 
    "Administrator", "Consultant", "Marketing", "Vertrieb"
]

# Function to extract relevant job keywords from a job posting
def extract_job_keywords(job_description, keyword_list, max_distance=3):
    extracted_keywords = []
    words = job_description.split()  # Split job description into words

    for word in words:
        for keyword in keyword_list:
            # Calculate Levenshtein distance between word and keyword
            distance = Levenshtein.distance(word, keyword)
            if distance <= max_distance:
                extracted_keywords.append(keyword)
                break  # Stop checking other keywords for this word
    
    return list(set(extracted_keywords))  # Return unique keywords

job_places=[
"Softwareentwickler (m/w/d) für dynamisches Unternehmen im Bereich Datenanalyse und Machine Learning",
"(Wirtschafts-)Mathematiker (m/w/d) für den Bereich | Hannover",       
"Mathematiker (m/w/d) im Versicherungsbereich - 30  | Stuttgart",      
"Mathematiker (m/w/d) im Risk Management mit Schwer | Garching bei München",
"Aktuar / Mathematiker Lebensversicherung | Fachkon | Essen      ",    
"Aktuar / Mathematiker Lebensversicherung | Fachkon | Mannheim ",   
"Aktuar / Mathematiker Lebensversicherung | Fachkon | Hamburg"]

for job in job_places: 
    extracted_keywords = extract_job_keywords(job, job_keywords)
    print("Extracted Keywords:", extracted_keywords)
```

```python
import os
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from git import Repo  # Installiere zuerst GitPython: pip install gitpython

# ChromeDriver Setup
def setup_driver():
    options = Options()
    options.add_argument("--remote-debugging-port=9222")  # Debugging Port für Jupyter
    options.add_argument("--start-maximized")
    service = Service(r'C:\Users\User\.wdm\drivers\chromedriver\win64\129.0.6668.100\chromedriver.exe')  # Ersetze durch den Pfad zu deinem ChromeDriver
    driver = webdriver.Chrome(service=service, options=options)
    return driver

# Funktion, um die Notebook-Datei zu extrahieren
def get_notebook_filename(driver):
    driver.get("http://localhost:8888/tree")  # Standard Jupyter-Notebook-URL
    time.sleep(3)

    # Suche die aktive Notebook-Datei
    try:
        active_file = driver.find_element(By.CLASS_NAME, "item_name").text
        print(f"Gefundene Notebook-Datei: {active_file}")
        return active_file
    except Exception as e:
        print(f"Fehler: {e}")
        return None

# Git-Funktion, um Datei zu pushen
def push_to_github(file_path, repo_path, commit_message="Automatischer Commit"):
    try:
        # Repository öffnen
        repo = Repo(repo_path)
        repo.git.add(file_path)
        repo.index.commit(commit_message)
        origin = repo.remote(name="origin")
        origin.push()
        print("Notebook erfolgreich gepusht!")
    except Exception as e:
        print(f"Fehler beim Push: {e}")

# Hauptprogramm
if __name__ == "__main__":
    # Starte Selenium
    driver = setup_driver()

    # Hole den Namen der aktiven Notebook-Datei
    notebook_file = get_notebook_filename(driver)

    if notebook_file:
        # Lokale Pfadeinstellungen
        notebook_path = os.path.join(os.getcwd(), notebook_file)  # Aktuelles Verzeichnis
        repo_path = "path/to/your/repo"  # Lokaler Pfad zum Repository

        # Datei zu GitHub pushen
        push_to_github(notebook_path, repo_path)
    else:
        print("Notebook-Datei konnte nicht gefunden werden.")

    # Beende Selenium
    driver.quit()
```
